# 线性情况

```python
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.r_[np.random.randn(20,2)-[2,2],np.random.randn(20,2)+[2,2]]#随机生成40个样本点   20个2维 20个2维
label=[0]*20+[1]*20#样本标签 20个标记为0,20个标记为1
plt.scatter(x[:,0],x[:,1],c=label)
plt.show()
```

```python
model=svm.SVC(kernel='linear')#核函数设置为线性函数
model.fit(x,label)
```

```python
print(model.support_vectors_)#查看支持向量
```

```python
print(model.n_support_)#查看支持向量的个数
```

```python
model.coef_#超平面
```

```python
model.intercept_
```

```python
#获取超平面
plt.scatter(x[:,0],x[:,1],c=label)
x_test=np.array([[-5],[5]])
d=-model.intercept_/model.coef_[0][1]
k=-model.coef_[0][0]/model.coef_[0][1]
y_test=d+k*x_test
plt.plot(x_test,y_test,'k')
plt.show()
```

```python
model.predict([[3,4]])#预测点（3,4）的类别
```

# 非线性的情况

```python
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D
x,y=datasets.make_circles(n_samples=1000,factor=.3,noise=.10)
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()#1000个样本点，不能线性划分
```

```python
z=x[:,0]**2+x[:,1]**2
```

```python
ax=plt.figure().add_subplot(111,projection='3d')
ax.scatter(x[:,0],x[:,1],z,c=y,s=10)
plt.show()
```
